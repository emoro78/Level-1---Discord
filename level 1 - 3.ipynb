```python
# Almacenar las credenciales de inicio de sesión
credentials = {'username': 'password'}

# Seguimiento del número de intentos de inicio de sesión
login_attempts = 0

# Guardar las citas
appointments = {}

# Definir los médicos y sus franjas horarias disponibles para cada especialidad
doctors = {
    'General Medicine': {'Dr. Smith': ['9:00', '10:00', '11:00'], 'Dr. Johnson': ['14:00', '15:00', '16:00'], 'Dr. Williams': ['9:00', '10:00', '11:00']},
    'Emergency Care': {'Dr. Brown': ['9:00', '10:00', '11:00'], 'Dr. Taylor': ['14:00', '15:00', '16:00'], 'Dr. Miller': ['9:00', '10:00', '11:00']},
    
    # Añadir otras especialidades y médicos
}

# Gestionar el proceso de inicio de sesión
def login():
    global login_attempts
    while login_attempts < 3:
        username = input("Enter your username: ")
        password = input("Enter your password: ")
        if credentials.get(username) == password:
            return True
        else:
            login_attempts += 1
            print("Invalid credentials. Please try again.")
    print("Account locked due to too many failed login attempts.")
    return False

# Visualizar las especialidades disponibles
def display_specialties():
    print("Available specialties:")
    for specialty in doctors.keys():
        print(specialty)

# Mostrar los médicos disponibles para una especialidad
def display_doctors(specialty):
    print(f"Available doctors for {specialty}:")
    for doctor in doctors[specialty].keys():
        print(doctor)

# Visualizar las franjas horarias disponibles para un médico en una especialidad
def display_time_slots(specialty, doctor):
    print(f"Available time slots for {doctor}:")
    for time_slot in doctors[specialty][doctor]:
        print(time_slot)

# Reservar una cita
def book_appointment(username):
    if len(appointments.get(username, [])) >= 3:
        print("You have reached the maximum limit for appointments.")
        return
    display_specialties()
    specialty = input("Enter your preferred specialty: ")
    display_doctors(specialty)
    doctor = input("Enter your preferred doctor: ")
    display_time_slots(specialty, doctor)
    time_slot = input("Enter your preferred time slot: ")
    if time_slot in doctors[specialty][doctor]:
        doctors[specialty][doctor].remove(time_slot)
        appointments.setdefault(username, []).append((specialty, doctor, time_slot))
        print("Appointment booked successfully.")
    else:
        print("Time slot not available. Please choose another time slot.")

# Gestionar el sistema de citas del hospital
def hospital_appointment_system():
    if login():
        while True:
            book_appointment('username')
            another_appointment = input("Do you want to book another appointment? (yes/no): ")
            if another_appointment.lower() != 'yes':
                break

# Iniciar el sistema de citas del hospital
hospital_appointment_system()
```
