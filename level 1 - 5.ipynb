```python
# Definir las variables para almacenar las credenciales de inicio de sesión
credentials = {'username': 'password'}

# Inicializar el contador de intentos de acceso
login_attempts = 0

# Inicializar el diccionario de reservas para almacenar las reservas de vuelos
bookings = {}

# Definir una lista de países disponibles
countries = ['Turkey', 'Greece', 'Lebanon', 'Spain', 'Portugal']

# Definir una lista de clases de vuelo disponibles
flight_classes = ['Economy', 'First Class']

# Definir una lista de tipos de comidas disponibles
meal_types = ['Regular', 'Vegetarian', 'Kosher']

# Definir una función para gestionar el proceso de inicio de sesión
def login():
    global login_attempts
    while login_attempts < 3:
        username = input("Enter your username: ")
        password = input("Enter your password: ")
        if credentials.get(username) == password:
            return True
        else:
            login_attempts += 1
            print("Invalid credentials. Please try again.")
    print("Account locked due to too many failed login attempts.")
    return False

# Definir una función para mostrar los países disponibles
def display_countries():
    print("Available countries:")
    for country in countries:
        print(country)

# Definir una función para mostrar las clases de vuelo disponibles
def display_flight_classes():
    print("Available flight classes:")
    for flight_class in flight_classes:
        print(flight_class)

# Definir una función para mostrar los tipos de comida disponibles
def display_meal_types():
    print("Available meal types:")
    for meal_type in meal_types:
        print(meal_type)

# Definir una función para gestionar el proceso de reserva de vuelos
def book_flight(username):
    display_countries()
    origin_country = input("Enter your origin country: ")
    destination_country = input("Enter your destination country: ")
    flight_date = input("Enter your flight date (YYYY-MM-DD): ")
    display_flight_classes()
    flight_class = input("Enter your preferred flight class: ")
    luggage = input("Do you want to check an additional piece of luggage into the hold? (yes/no): ")
    display_meal_types()
    meal_type = input("Enter your preferred meal type: ")
    name = input("Enter your name: ")
    passport = input("Enter your passport number: ")
    print(f"\nSummary of your choices:\nOrigin country: {origin_country}\nDestination country: {destination_country}\nFlight date: {flight_date}\nFlight class: {flight_class}\nAdditional luggage: {luggage}\nMeal type: {meal_type}\nName: {name}\nPassport number: {passport}")
    confirm = input("\nDo you confirm the reservation? (yes/no): ")
    if confirm.lower() == 'yes':
        bookings.setdefault(username, []).append((origin_country, destination_country, flight_date, flight_class, luggage, meal_type, name, passport))
        print("Reservation confirmed.")
    else:
        print("Reservation cancelled.")

# Definir la función principal para ejecutar el sistema de reserva de vuelos
def flight_booking_system():
    if login():
        while True:
            book_flight('username')
            another_booking = input("Do you want to make another booking? (yes/no): ")
            if another_booking.lower() != 'yes':
                break

# Ejecutar el sistema de reserva de vuelos
flight_booking_system()
```
