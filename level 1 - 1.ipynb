```python
# Definios las variables con las playeras y sus estadisticas
players = {
    8: {'name': 'Bruno Fernandes', 'goals': 5, 'speed': 6, 'assists': 9, 'passing_accuracy': 10, 'defensive_involvements': 3},
    11: {'name': 'Rasmus Hojlund', 'goals': 12, 'speed': 8, 'assists': 2, 'passing_accuracy': 6, 'defensive_involvements': 2},
    5: {'name': 'Harry Maguire', 'goals': 1, 'speed': 5, 'assists': 1, 'passing_accuracy': 7, 'defensive_involvements': 9},
    17: {'name': 'Alejandro Garnacho', 'goals': 8, 'speed': 7, 'assists': 8, 'passing_accuracy': 6, 'defensive_involvements': 0},
    7: {'name': 'Mason Mount', 'goals': 2, 'speed': 6, 'assists': 4, 'passing_accuracy': 8, 'defensive_involvements': 1}
}

# Mostrar las estadisticas de un jugador
def player_review(jersey_number):
    player = players[jersey_number]
    print(f"Player: {player['name']}")
    print(f"Goals: {player['goals']}")
    print(f"Speed: {player['speed']}")
    print(f"Assists: {player['assists']}")
    print(f"Passing Accuracy: {player['passing_accuracy']}")
    print(f"Defensive Involvements: {player['defensive_involvements']}")

# Comparar las estadisticas de dos jugadores
def compare_players(jersey_number1, jersey_number2):
    player1 = players[jersey_number1]
    player2 = players[jersey_number2]
    print(f"{player1['name']} vs {player2['name']}")
    for stat in ['goals', 'speed', 'assists', 'passing_accuracy', 'defensive_involvements']:
        print(f"{stat}: {player1[stat]} vs {player2[stat]}")

# Function to identify the player with the highest value for a given statistic
def identify_top_player(stat):
    top_player = max(players.values(), key=lambda player: player[stat])
    print(f"Top player in {stat}: {top_player['name']} with {top_player[stat]} points")

# Identificar al jugador con el valor más alto para una estadística dada
def player_statistics():
    while True:
        print("1. Player Review")
        print("2. Compare Two Players")
        print("3. Identify Fastest Player")
        print("4. Identify Top Goal Scorer")
        print("5. Identify Player with Most Assists")
        print("6. Identify Player with Highest Passing Accuracy")
        print("7. Identify Player with Most Defensive Involvements")
        print("8. Exit")
        choice = int(input("Enter your choice: "))
        if choice == 1:
            jersey_number = int(input("Enter the jersey number of the player: "))
            player_review(jersey_number)
        elif choice == 2:
            jersey_number1 = int(input("Enter the jersey number of the first player: "))
            jersey_number2 = int(input("Enter the jersey number of the second player: "))
            compare_players(jersey_number1, jersey_number2)
        elif choice == 3:
            identify_top_player('speed')
        elif choice == 4:
            identify_top_player('goals')
        elif choice == 5:
            identify_top_player('assists')
        elif choice == 6:
            identify_top_player('passing_accuracy')
        elif choice == 7:
            identify_top_player('defensive_involvements')
        elif choice == 8:
            break

# Llamar a la función player_statistics para iniciar el programa
player_statistics()
```
